## Exercício 4

Qual das matrizes abaixo aplicaria uma transformação de contração?

```python
# Matriz a - contração
x = np.random.uniform(-0.5, 0.5, size=(2, 600))
A = np.array([ [0.34, 0], [0, 0.94] ])
y = A @ x

plt.figure()
plt.scatter(x[0,:], x[1,:], s=1, c='r')
plt.scatter(y[0,:], y[1,:], s=1, c='b')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.grid()
plt.show()
```

```python
# Matriz b - cisalhamento
x = np.random.uniform(-0.5, 0.5, size=(2, 600))
A = np.array([ [1, 0.34], [0.94, 1] ])
y = A @ x

plt.figure()
plt.scatter(x[0,:], x[1,:], s=1, c='r')
plt.scatter(y[0,:], y[1,:], s=1, c='b')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.grid()
plt.show()
```

```python
# Matriz c - rotação
x = np.random.uniform(-0.5, 0.5, size=(2, 600))
A = np.array([ [0.94, -0.34], [0.34, 0.94] ])
y = A @ x

plt.figure()
plt.scatter(x[0,:], x[1,:], s=1, c='r')
plt.scatter(y[0,:], y[1,:], s=1, c='b')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.grid()
plt.show()
```

## Exercício 5

Qual matriz aplicaria a transformação na nuvem de pontos à esquerda de modo que ela fique como a nuvem de pontos à direita?

```python
# Matriz de rotação
x = np.random.uniform(-0.5, 0.5, size=(2, 600))
A = np.array([ [0.5, -0.8], [0.8, 0.5] ])
y = A @ x

plt.figure()
plt.scatter(x[0,:], x[1,:], s=1, c='r')
plt.scatter(y[0,:], y[1,:], s=1, c='b')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.grid()
plt.show()
```
